```python
import tensorflow as tf
import json
import numpy as np
from tensorflow import keras as tfk
import os
from pymongo import MongoClient
import pandas as pd
```

```python
convos = ['conversation1', 'conversation2', 'conversation3']
vocons = []
for y in convos:
#     with open(f'conversations_kaggle/{y}.txt', 'r') as unoconvo:
#         vocons.append(unoconvo.read())
    vocons.append(pd.read_csv('conversations_kaggle/'+str(y)+'.txt', delimiter='\n\n', header=None, engine='python'))
```

```python
for y, x in enumerate(vocons):
    x.to_csv('conversations_kaggle/'+convos[y]+'.csv', index=True)
```
